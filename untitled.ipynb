```python
import numpy as np

class HfO2_Simulator:
    def __init__(self, lp=2.4, ld=-2.1, bits=6):
        # NeuroSim 3.0 표준 HfO2 파라미터 (단위: Siemens)
        self.g_max = 100e-6 
        self.g_min = 1e-6
        self.lp = lp  # 증가 시 비선형성
        self.ld = ld  # 감소 시 비선형성
        self.levels = 2 ** bits # 64 levels
        
        # 64개의 불연속적인 상태(State) 생성
        self.states = np.linspace(0, 1, self.levels)
        self.lut_p = self._calc_g(self.states, self.lp)
        self.lut_d = self._calc_g(self.states, self.ld)

    def _calc_g(self, p, alpha):
        if alpha == 0: return self.g_min + (self.g_max - self.g_min) * p
        return self.g_min + (self.g_max - self.g_min) * (1 - np.exp(-alpha * p)) / (1 - np.exp(-alpha))

    def apply_device_constraints(self, weight):
        """수학적 가중치(0~1)를 실제 소자가 가질 수 있는 전도도로 변환"""
        # 1. 0~1 사이로 정규화된 가중치를 전도도 범위로 확장
        target_g = self.g_min + (self.g_max - self.g_min) * weight
        
        # 2. 64단계 중 가장 가까운 물리적 상태 찾기 (Resolution 한계 반영)
        idx = np.argmin(np.abs(self.lut_p - target_g))
        real_g = self.lut_p[idx]
        
        # 3. 다시 가중치 형태(0~1)로 복원하여 모델에 적용
        return (real_g - self.g_min) / (self.g_max - self.g_min)
```

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from spikingjelly.activation_based import neuron, layer, functional, surrogate

# 1. NeuroSim HfO2 파라미터 (Cell.cpp 수치 반영)
class HfO2_RRAM:
    NL_LTP = 2.4
    NL_LTD = -4.88
    STEPS = 100
    SIGMA = 0.035

# 2. 고도화된 차동 시냅스 엔진 (초기 발화율 확보)
class NeuroSim_Synapse:
    def __init__(self, shape, device):
        self.device = device
        # [핵심 수정] 초기 가중치가 0이 되지 않도록 G_pos와 G_neg에 차이를 둠
        # 알고리즘상의 Kaiming Normal 분포를 물리적 공간([0.1, 0.9])에 매핑
        init_w = torch.randn(shape, device=device) * 0.1
        self.g_pos = torch.clamp(0.5 + (init_w / 2.0), 0.1, 0.9)
        self.g_neg = torch.clamp(0.5 - (init_w / 2.0), 0.1, 0.9)

    def get_weight(self):
        # 가중치 스케일을 키워 뉴런이 발화할 수 있게 함 (Scale factor 5.0)
        return (self.g_pos - self.g_neg) * 5.0

    def update_physics(self, delta_w):
        with torch.no_grad():
            # 변화량에 따라 업데이트 방향 결정 (임계값 완화 0.001)
            mask_p = delta_w > 0.001
            mask_d = delta_w < -0.001

            # NeuroSim NonlinearWeight 공식 적용
            if mask_p.any():
                self.g_pos[mask_p] += (1.0 / HfO2_RRAM.STEPS) * torch.exp(-HfO2_RRAM.NL_LTP * self.g_pos[mask_p])
            if mask_d.any():
                self.g_neg[mask_d] += (1.0 / HfO2_RRAM.STEPS) * torch.exp(-HfO2_RRAM.NL_LTP * self.g_neg[mask_d])

            # 노이즈 및 클리핑
            noise = torch.randn_like(self.g_pos) * HfO2_RRAM.SIGMA
            self.g_pos = torch.clamp(self.g_pos + noise, 0.01, 0.99)
            self.g_neg = torch.clamp(self.g_neg + noise, 0.01, 0.99)

# 3. 모델 정의
class RealisticSNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            layer.Flatten(),
            layer.Linear(784, 100, bias=False),
            neuron.LIFNode(tau=2.0, surrogate_function=surrogate.Sigmoid(), detach_reset=True),
            layer.Linear(100, 10, bias=False),
            neuron.LIFNode(tau=2.0, surrogate_function=surrogate.Sigmoid(), detach_reset=True)
        )
    def forward(self, x): return self.model(x)

# 4. 분석 실행
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# 데이터 정규화 강화 (0.5 평균)
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])
train_loader = DataLoader(datasets.MNIST('./data', train=True, download=True, transform=transform), batch_size=128, shuffle=True)
test_loader = DataLoader(datasets.MNIST('./data', train=False, transform=transform), batch_size=128)

model = RealisticSNN().to(device)
syn_fc1 = NeuroSim_Synapse(model.model[1].weight.shape, device)
syn_fc2 = NeuroSim_Synapse(model.model[3].weight.shape, device)

optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

acc_history = []
T = 8

print(f"Starting Final Analysis on {device}...")

for epoch in range(5):
    model.train()
    for img, label in train_loader:
        img, label = img.to(device), label.to(device)
        
        # 1. 현재 물리적 가중치 동기화
        model.model[1].weight.data = syn_fc1.get_weight()
        model.model[3].weight.data = syn_fc2.get_weight()
        
        # 2. 순전파 및 역전파
        optimizer.zero_grad()
        out_fr = 0
        for t in range(T): out_fr += model(img)
        loss = criterion(out_fr / T, label)
        loss.backward()
        
        # 3. 가상 업데이트 수행
        optimizer.step()
        
        # 4. 가상 가중치와의 차이를 물리적 펄스로 반영
        with torch.no_grad():
            syn_fc1.update_physics(model.model[1].weight.data - syn_fc1.get_weight())
            syn_fc2.update_physics(model.model[3].weight.data - syn_fc2.get_weight())
        
        functional.reset_net(model)

    # 평가
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for img, label in test_loader:
            img, label = img.to(device), label.to(device)
            out_fr = 0
            for t in range(T): out_fr += model(img)
            correct += (out_fr.argmax(1) == label).sum().item()
            total += label.size(0)
            functional.reset_net(model)
    
    acc = correct / total
    acc_history.append(acc)
    print(f"Epoch {epoch+1}: Accuracy = {acc:.4f}")

plt.plot(range(1, 6), acc_history, 'bo-', label='HfO2 Differential RRAM')
plt.ylabel('Accuracy'); plt.xlabel('Epoch'); plt.legend(); plt.show()
```

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from spikingjelly.activation_based import neuron, layer, functional, surrogate

# 1. HfO2 RRAM 물리 파라미터 (고정)
class HfO2_Physics:
    NL_LTP = 2.4
    NL_LTD = -4.88
    STEPS = 100
    SIGMA = 0.035

# 2. 장기 수렴을 위해 '센터링'이 강화된 시냅스 엔진
class Robust_Synapse:
    def __init__(self, shape, device):
        self.device = device
        # 검증된 초기화 로직 유지
        self.g_pos = torch.full(shape, 0.55, device=device) + torch.randn(shape, device=device) * 0.03
        self.g_neg = torch.full(shape, 0.45, device=device) + torch.randn(shape, device=device) * 0.03
        
        # [조정] LTD 보정을 더 보수적으로 (0.25)
        self.comp_factor = 0.25 
        # 가중치 포화 방지를 위한 미세 복원력 (Centering)
        self.leak = 0.0005 

    def get_weight(self):
        # 검증된 신호 세기 유지
        return (self.g_pos - self.g_neg) * 6.5

    def update_physics(self, delta_w):
        with torch.no_grad():
            # [추가] 그레디언트 클리핑: 급격한 가중치 변화 방지
            delta_w = torch.clamp(delta_w, -0.05, 0.05)
            
            mask_p = delta_w > 0.001
            mask_d = delta_w < -0.001

            # LTP/LTD 물리 업데이트
            if mask_p.any():
                self.g_pos[mask_p] += (1.0 / HfO2_Physics.STEPS) * torch.exp(-HfO2_Physics.NL_LTP * self.g_pos[mask_p])
            if mask_d.any():
                # g_neg를 키워 전체 가중치를 낮춤 (LTD 보정 적용)
                self.g_neg[mask_d] += (self.comp_factor / HfO2_Physics.STEPS) * torch.exp(-HfO2_Physics.NL_LTP * self.g_neg[mask_d])

            # [핵심] 가중치 센터링: 0이나 1로 달라붙는 것을 방지
            self.g_pos += (0.5 - self.g_pos) * self.leak
            self.g_neg += (0.5 - self.g_neg) * self.leak

            # 물리적 노이즈 및 가중치 활성 영역(0.05 ~ 0.95) 보호
            noise = torch.randn_like(self.g_pos) * HfO2_Physics.SIGMA
            self.g_pos = torch.clamp(self.g_pos + noise, 0.05, 0.95)
            self.g_neg = torch.clamp(self.g_neg + noise, 0.05, 0.95)

# 3. 모델 정의
class RealisticSNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            layer.Flatten(),
            layer.Linear(784, 100, bias=False),
            neuron.LIFNode(tau=2.0, surrogate_function=surrogate.Sigmoid(), detach_reset=True),
            layer.Linear(100, 10, bias=False),
            neuron.LIFNode(tau=2.0, surrogate_function=surrogate.Sigmoid(), detach_reset=True)
        )
    def forward(self, x): return self.model(x)

# 4. 분석 환경 및 데이터 설정
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.1307,), (0.3081,))])
train_loader = DataLoader(datasets.MNIST('./data', train=True, download=True, transform=transform), batch_size=128, shuffle=True)
test_loader = DataLoader(datasets.MNIST('./data', train=False, transform=transform), batch_size=128)

model = RealisticSNN().to(device)
syn1 = Robust_Synapse(model.model[1].weight.shape, device)
syn2 = Robust_Synapse(model.model[3].weight.shape, device)

# 학습률을 0.0008로 소폭 낮추어 안정성 강화
optimizer = torch.optim.Adam(model.parameters(), lr=0.0008)
# 코사인 스케줄러로 에포크 후반부 업데이트를 더 정교하게 제어
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=10)
criterion = nn.CrossEntropyLoss()

acc_history = []
T = 8 

print(f"Starting Robust Stability Analysis on {device}...")

# 5. 메인 학습 루프
for epoch in range(10):
    model.train()
    for img, label in train_loader:
        img, label = img.to(device), label.to(device)
        model.model[1].weight.data = syn1.get_weight()
        model.model[3].weight.data = syn2.get_weight()
        
        optimizer.zero_grad()
        out_fr = 0
        for t in range(T): out_fr += model(img)
        loss = criterion(out_fr / T, label)
        loss.backward()
        optimizer.step()
        
        with torch.no_grad():
            syn1.update_physics(model.model[1].weight.data - syn1.get_weight())
            syn2.update_physics(model.model[3].weight.data - syn2.get_weight())
        
        functional.reset_net(model)
    
    scheduler.step()
    
    # 평가
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for img, label in test_loader:
            img, label = img.to(device), label.to(device)
            out_fr = 0
            for t in range(T): out_fr += model(img)
            correct += (out_fr.argmax(1) == label).sum().item()
            total += label.size(0)
            functional.reset_net(model)
    
    acc = correct / total
    acc_history.append(acc)
    print(f"Epoch {epoch+1}: Accuracy = {acc:.4f}, LR = {optimizer.param_groups[0]['lr']:.6f}")

# 결과 시각화
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), acc_history, 'mo-', label='Robust HfO2 SNN')
plt.title('Final Optimized Accuracy: Preventing Epoch Crash')
plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.grid(True); plt.show()
```

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from spikingjelly.activation_based import neuron, layer, functional, surrogate

# 1. HfO2 RRAM 물리 특성 설정
class HfO2_Device:
    NL_LTP = 2.4

# 2. 해상도와 레이어 스케일이 독립 제어되는 HAT 레이어
class RRAM_Linear_HAT(nn.Linear):
    def __init__(self, in_features, out_features, res_steps=100):
        super().__init__(in_features, out_features, bias=False)
        self.res_steps = res_steps
        
        # [핵심 수정] 가중치를 소자 가동 범위인 [-0.8, 0.8]에 고르게 배치하여 
        # 어떤 해상도에서도 균일한 격자 샘플링이 일어나도록 유도
        nn.init.uniform_(self.weight, -0.8, 0.8)
        
        # [핵심 수정] 레이어 크기에 맞는 최적의 입력 에너지를 동적으로 계산 (Kaiming 기반 확장)
        # LIF 뉴런이 죽지 않고 가장 민감하게 반응하는 Firing Gain(4.5) 적용
        self.dynamic_scale = np.sqrt(2.0 / in_features) * 4.5

    def apply_physics(self, w):
        w_clipped = torch.clamp(w, -1.0, 1.0)
        
        # 차동 구조 모사 (Positive / Negative 소자 분리)
        w_pos = torch.clamp(w_clipped, min=0)
        w_neg = torch.clamp(-w_clipped, min=0)
        
        # STE(Straight-Through Estimator) 양자화: 정규화된 공간에 해상도 정밀 적용
        w_pos_q = w_pos + (torch.round(w_pos * self.res_steps) / self.res_steps - w_pos).detach()
        w_neg_q = w_neg + (torch.round(w_neg * self.res_steps) / self.res_steps - w_neg).detach()
        
        # HfO2 비선형성 곡선 투영
        w_pos_phys = (1 - torch.exp(-HfO2_Device.NL_LTP * w_pos_q)) / (1 - np.exp(-HfO2_Device.NL_LTP))
        w_neg_phys = (1 - torch.exp(-HfO2_Device.NL_LTP * w_neg_q)) / (1 - np.exp(-HfO2_Device.NL_LTP))
        
        # 물리 연산 종료 후, 레이어별 밸런스를 맞추기 위한 동적 스케일 적용
        return (w_pos_phys - w_neg_phys) * self.dynamic_scale

    def forward(self, x):
        p_weight = self.apply_physics(self.weight)
        return nn.functional.linear(x, p_weight)

# 3. SNN 모델 정의
class HfO2_Quant_SNN(nn.Module):
    def __init__(self, res_steps):
        super().__init__()
        self.model = nn.Sequential(
            layer.Flatten(),
            RRAM_Linear_HAT(784, 128, res_steps=res_steps),
            neuron.LIFNode(tau=2.0, surrogate_function=surrogate.Sigmoid(), detach_reset=True),
            RRAM_Linear_HAT(128, 10, res_steps=res_steps),
            neuron.LIFNode(tau=2.0, surrogate_function=surrogate.Sigmoid(), detach_reset=True)
        )
    def forward(self, x): return self.model(x)

num_operator = 30

# 4. 해상도별 정량 실험 제어 함수
def run_resolution_analysis(res_list):
    from torchvision import datasets, transforms
    from torch.utils.data import DataLoader
    
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.1307,), (0.3081,))])
    train_loader = DataLoader(datasets.MNIST('./data', train=True, download=True, transform=transform), batch_size=128, shuffle=True)
    test_loader = DataLoader(datasets.MNIST('./data', train=False, transform=transform), batch_size=128)

    all_results = {}

    for res in res_list:
        print(f"\n--- 분석 시작: Resolution {res} Steps ---")
        model = HfO2_Quant_SNN(res_steps=res).to(device)
        optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
        criterion = nn.CrossEntropyLoss()
        
        history = []
        for epoch in range(num_operator):
            model.train()
            for img, label in train_loader:
                img, label = img.to(device), label.to(device)
                optimizer.zero_grad()
                
                out_fr = 0
                for t in range(8): out_fr += model(img)
                loss = criterion(out_fr / 8, label)
                loss.backward()
                
                optimizer.step()
                functional.reset_net(model)
            
            # 평가
            model.eval()
            correct, total = 0, 0
            with torch.no_grad():
                for img, label in test_loader:
                    img, label = img.to(device), label.to(device)
                    out_fr = 0
                    for t in range(8): out_fr += model(img)
                    correct += (out_fr.argmax(1) == label).sum().item()
                    total += label.size(0)
                    functional.reset_net(model)
            
            acc = correct / total
            history.append(acc)
            print(f"Epoch {epoch+1}: 정확도 = {acc:.4f}")
        
        all_results[res] = history
    return all_results

# 5. 최종 실험 제어 (8, 16, 64, 256 전 영역 완전 통제 비교)
resolutions = [8, 16, 64, 256]
analysis_data = run_resolution_analysis(resolutions)

plt.figure(figsize=(10, 6))
for res, history in analysis_data.items():
    plt.plot(range(1, num_operator+1), history, marker='o', label=f'Resolution: {res} steps')
plt.title('Quantitative Impact of RRAM Resolution on SNN Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from spikingjelly.activation_based import neuron, layer, functional, surrogate

class HfO2_Device:
    NL_LTP = 2.4

# 해상도와 레이어 스케일이 독립 제어되는 HAT 레이어
class RRAM_Linear_HAT(nn.Linear):
    def __init__(self, in_features, out_features, res_steps=100):
        super().__init__(in_features, out_features, bias=False)
        self.res_steps = res_steps
        
        nn.init.uniform_(self.weight, -0.8, 0.8)
        self.dynamic_scale = np.sqrt(2.0 / in_features) * 4.5

        import math
        self.norm_factor = 1.0 - math.exp(-HfO2_Device.NL_LTP)

    def apply_physics(self, w):
        w_clipped = torch.clamp(w, -1.0, 1.0)
        
        w_pos = torch.clamp(w_clipped, min=0)
        w_neg = torch.clamp(-w_clipped, min=0)
        
        w_pos_q = w_pos + (torch.round(w_pos * self.res_steps) / self.res_steps - w_pos).detach()
        w_neg_q = w_neg + (torch.round(w_neg * self.res_steps) / self.res_steps - w_neg).detach()
        
        w_pos_phys = (1.0 - torch.exp(-HfO2_Device.NL_LTP * w_pos_q)) / self.norm_factor
        w_neg_phys = (1.0 - torch.exp(-HfO2_Device.NL_LTP * w_neg_q)) / self.norm_factor
        
        return (w_pos_phys - w_neg_phys) * self.dynamic_scale

    def forward(self, x):
        p_weight = self.apply_physics(self.weight)
        return nn.functional.linear(x, p_weight)

# SNN 모델 정의
class HfO2_Quant_SNN(nn.Module):
    def __init__(self, res_steps):
        super().__init__()
        self.model = nn.Sequential(
            layer.Flatten(),
            RRAM_Linear_HAT(784, 128, res_steps=res_steps),
            neuron.LIFNode(tau=2.0, surrogate_function=surrogate.Sigmoid(), detach_reset=True),
            RRAM_Linear_HAT(128, 10, res_steps=res_steps),
            neuron.LIFNode(tau=2.0, surrogate_function=surrogate.Sigmoid(), detach_reset=True)
        )
    def forward(self, x): return self.model(x)

num_operator = 30

# 해상도별 정량 실험 제어 함수
def run_resolution_analysis(res_list):
    from torchvision import datasets, transforms
    from torch.utils.data import DataLoader
    
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.1307,), (0.3081,))])
    train_loader = DataLoader(datasets.MNIST('./data', train=True, download=True, transform=transform), batch_size=128, shuffle=True)
    test_loader = DataLoader(datasets.MNIST('./data', train=False, transform=transform), batch_size=128)

    all_results = {}

    for res in res_list:
        print(f"\n--- 분석 시작: Resolution {res} Steps ---")
        model = HfO2_Quant_SNN(res_steps=res).to(device)
        optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
        criterion = nn.CrossEntropyLoss()
        
        history = []
        for epoch in range(num_operator):
            model.train()
            for img, label in train_loader:
                img, label = img.to(device), label.to(device)
                optimizer.zero_grad()
                
                out_fr = 0
                for t in range(8): out_fr += model(img)
                loss = criterion(out_fr / 8, label)
                loss.backward()
                
                optimizer.step()
                functional.reset_net(model)
            
            # 평가
            model.eval()
            correct, total = 0, 0
            with torch.no_grad():
                for img, label in test_loader:
                    img, label = img.to(device), label.to(device)
                    out_fr = 0
                    for t in range(8): out_fr += model(img)
                    correct += (out_fr.argmax(1) == label).sum().item()
                    total += label.size(0)
                    functional.reset_net(model)
            
            acc = correct / total
            history.append(acc)
            print(f"Epoch {epoch+1}: 정확도 = {acc:.4f}")
        
        all_results[res] = history
    return all_results

# 최종 실험 제어 (8, 16, 64, 256 전 영역 완전 통제 비교)
resolutions = [8, 16, 64, 256]
analysis_data = run_resolution_analysis(resolutions)

plt.figure(figsize=(10, 6))
for res, history in analysis_data.items():
    plt.plot(range(1, num_operator+1), history, marker='o', label=f'Resolution: {res} steps')
plt.title('Quantitative Impact of RRAM Resolution on SNN Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import re
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

# 1. 설정값
log_file_path = "log/output_t4_ep100.log"  # 다운로드한 로그 파일 이름
target_epochs = 90  # 💡 관찰하고 싶은 에포크까지만 자르기 (예: 수렴이 일찍 끝난 40에포크까지만)
target_resolution = 16
baselines = ['Ideal_FP32', 'Linear_Quant', 'HfO2_Physics', 'Noise_Aware']

# 2. 데이터를 담을 딕셔너리 생성
# 구조: ablation_data[mode][epoch] = [acc1, acc2, acc3, acc4, acc5]
ablation_data = defaultdict(lambda: defaultdict(list))

# 3. 정규표현식(Regex)을 이용해 로그 텍스트에서 데이터 추출
# 파이썬 코드의 print(f"[{mode} | Seed {seed}] Epoch {epoch+1}/{epochs} - Acc: {correct/total:.4f}") 구조를 정확히 캡처합니다.
log_pattern = re.compile(r"\[(.*?) \| Seed (\d+)\] Epoch (\d+)/\d+ - Acc: ([\d\.]+)")

with open(log_file_path, "r", encoding="utf-8") as f:
    for line in f:
        match = log_pattern.search(line)
        if match:
            mode = match.group(1)
            # epoch는 1부터 시작하게 찍혔으므로, 0-index 배열을 위해 1을 빼줍니다.
            epoch = int(match.group(3)) - 1 
            acc = float(match.group(4))
            
            # 지정한 target_epochs 미만의 데이터만 수집
            if epoch < target_epochs:
                ablation_data[mode][epoch].append(acc)

# 4. 수집된 데이터를 바탕으로 그래프 그리기
plt.figure(figsize=(12, 7))

for mode in baselines:
    if mode not in ablation_data:
        continue  # 해당 모드의 로그가 없으면 건너뜀
        
    # 데이터가 존재하는 에포크 구간만 정렬하여 추출
    epochs_range = sorted(list(ablation_data[mode].keys()))
    
    if not epochs_range:
        continue
        
    means = [np.mean(ablation_data[mode][e]) for e in epochs_range]
    stds = [np.std(ablation_data[mode][e]) for e in epochs_range]
    
    # 꺾은선 그래프 및 오차 범위(Shaded Area) 시각화
    plt.plot(epochs_range, means, label=mode, linewidth=2)
    plt.fill_between(epochs_range, np.array(means) - np.array(stds), np.array(means) + np.array(stds), alpha=0.2)

# 5. 최종 그래프 꾸미기 및 저장
plt.title(f'Ablation Study Zoom-in (First {target_epochs} Epochs)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.tight_layout()

save_name = f'zoomed_ablation_{target_epochs}ep.png'
plt.show()
print(f"✅ 그래프 추출 완료: {save_name} 파일이 생성되었습니다.")
```

```python
import os
import glob
import re
import numpy as np
from collections import defaultdict

def analyze_ep100_logs_per_file(log_dir="."):
    # 1. 파일 찾기
    log_files = glob.glob(os.path.join(log_dir, "log/output_*_ep100.log"))
    
    if not log_files:
        print(f"'{log_dir}' 경로에 output_***_ep100.log 파일이 없습니다.")
        return

    print(f"총 {len(log_files)}개의 로그 파일을 '개별적'으로 분석합니다.\n")

    pattern = re.compile(r"\[(.*?)\s*\|\s*Seed\s*(\d+)\]\s*Epoch\s*(\d+)/\d+\s*-\s*Acc:\s*([\d\.]+)")
    target_modes = ['Ideal_FP32', 'Linear_Quant', 'HfO2_Physics', 'Noise_Aware']
    target_epoch = 1
    final_epoch = 100

    # 2. 각 파일별로 루프를 돌며 독립된 딕셔너리(data) 생성
    for file_path in log_files:
        file_name = os.path.basename(file_path)
        print("=" * 70)
        print(f" 📂 [분석 대상 파일]: {file_name}")
        print("=" * 70)
        
        # 파일마다 데이터를 새로 초기화 (섞임 방지)
        data = defaultdict(lambda: defaultdict(list))
        
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f:
                match = pattern.search(line)
                if match:
                    mode = match.group(1).strip()
                    seed = int(match.group(2))
                    epoch = int(match.group(3))
                    acc = float(match.group(4))
                    data[mode][epoch].append(acc)

        # ---------------------------------------------------------
        # [분석 1] 초기(Epoch 1) 정량적 격차 분석
        # ---------------------------------------------------------
        stats = {}
        for mode in target_modes:
            if mode in data and target_epoch in data[mode]:
                acc_list = data[mode][target_epoch]
                stats[mode] = {'mean': np.mean(acc_list), 'std': np.std(acc_list)}
            else:
                stats[mode] = {'mean': 0.0, 'std': 0.0}

        fp32_mean = stats['Ideal_FP32']['mean']
        quant_mean = stats['Linear_Quant']['mean']
        physics_mean = stats['HfO2_Physics']['mean']

        print(f"\n▶ [Epoch {target_epoch}] 정량적 오차 분석 (초기 충격량)")
        if fp32_mean > 0:
            res_penalty = fp32_mean - quant_mean
            asym_penalty = quant_mean - physics_mean
            print(f"  - [기준] Ideal_FP32 평균     : {fp32_mean*100:.2f}%")
            print(f"  - [격차] 해상도 한계 오차    : {res_penalty*100:+.2f}%p (FP32 - Quant)")
            print(f"  - [격차] 비대칭성 오차       : {asym_penalty*100:+.2f}%p (Quant - Physics)")
            print(f"  - [불안정성] FP32(σ {stats['Ideal_FP32']['std']*100:.2f}%) vs Physics(σ {stats['HfO2_Physics']['std']*100:.2f}%)")
        else:
            print("  - 데이터 없음")

        # ---------------------------------------------------------
        # [분석 2] 최종(Epoch 100) 수렴 분석
        # ---------------------------------------------------------
        print(f"\n▶ [Epoch {final_epoch}] 최종 수렴 및 분산(σ) 평탄화 결과")
        has_final_data = any(final_epoch in data[mode] for mode in target_modes)
        
        if has_final_data:
            for mode in target_modes:
                if mode in data and final_epoch in data[mode]:
                    acc_list = data[mode][final_epoch]
                    mean_val = np.mean(acc_list)
                    std_val = np.std(acc_list)
                    print(f"  - {mode:15s} | 평균: {mean_val*100:.2f}% | 표준편차(에러바 두께): {std_val*100:.4f}%")
        else:
            print("  - 아직 Epoch 100 데이터가 기록되지 않았습니다.")
        print("\n")

if __name__ == "__main__":
    analyze_ep100_logs_per_file(".")
```
